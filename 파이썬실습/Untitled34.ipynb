```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import pandas as pd
data=pd.read_csv('40개변수대한항공.csv',engine='python')
del data['Unnamed: 0']
data.head()
```

```python
f,ax=plt.subplots(1,3,figsize=(20,8))
sns.distplot(data[data['low']].close,ax=ax[0])
ax[0].set_title('Fares in Pclass 1')
sns.distplot(data[data['high']].close,ax=ax[1])
ax[1].set_title('Fares in Pclass 2')
sns.distplot(data['volume'].close,ax=ax[2])
ax[2].set_title('Fares in Pclass 3')
plt.show()
```
