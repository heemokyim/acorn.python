```python
import pandas as pd
df=pd.read_csv('오뚜기40개변수추출.csv',engine='python')
```

```python
del df['Unnamed: 0']
df.head()
```

```python
df['yn']=df['close'] - df['beforeclose']
df.head()
```

```python
df.drop(['close'], inplace=True, axis=1)
```

```python
def stockflow(x):
    if x >= 0:
        return 1
    else:
        return 0
df['yn'] = df.yn.map(lambda x: stockflow(x))
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:39], df["yn"], test_size=0.33, random_state=42)
```

```python
import pandas
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

seed = 7
model = LogisticRegression().fit(X_train, y_train)

predicted = model.predict(X_test)
report = classification_report(y_test, predicted)
print(report)
matrix = confusion_matrix(y_test, predicted)
print(matrix)
```
