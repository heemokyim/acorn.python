```python
import pandas as pd
df = pd.read_csv('오뚜기.csv', engine='python')
df.head()
y = df['close']
```

```python
del df['date']
#del df['close']
df=df.fillna(0)
df.head()
```

```python
import sklearn
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBRegressor
from xgboost import plot_importance
from matplotlib import pyplot
scaler = MinMaxScaler(feature_range=(-1,1))
%matplotlib inline
df.loc[:,:'j'] = scaler.fit_transform(df.loc[:,:'j'])


X = df.loc[:,:'j']
#y = df['close']

import numpy as np
model = XGBRegressor().fit(X,y)
plot_importance(model)
pyplot.show()
```

```python
a=pd.DataFrame(model.feature_importances_)
b=X.columns.values
a['name']=b
a['importance']=a.iloc[:,0]
a = a.sort_values(['importance'], ascending=[False])
a = a.reset_index(drop=True)
ddugifeature=a.iloc[:40,:]
ddugifeature.name.values

ddugifeature.head()
```
